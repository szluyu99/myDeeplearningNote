#### sigmoid
$$h(x) = \frac{1}{1 + exp(-x)}$$

#### sigmoid

```python
import numpy as np
import matplotlib.pyplot as plt

# sigmoid
def sigmoid(x):
    return 1/(1 + np.exp(-x))

x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)

plt.plot(x, y)
plt.ylim(-0.1, 1.1) # y
```

#### sigmoid

```python
x = np.arange(-5.0, 5.0, 0.1)
y1 = step_function(x) # 
y2 = sigmoid(x) # sigmoid

plt.plot(x,y1,"--", label="step_function(x)")
plt.plot(x,y2,"-", label="sigmoid(x)")
plt.legend()
```

#### ReLU
sigmoid  **ReLU**(Rectified Linear Unit)  

$$
h(x)=\left\{
\begin{aligned}
x \  (x > 0) \\
0 \  (x \leqslant 0)\\
\end{aligned}
\right.
$$

```python
import numpy as np
import matplotlib.pyplot as plt

def relu(x): 
    return np.maximum(0, x) # maximun

x = np.arange(-6.0, 6.0, 1.0)
y = relu(x)

plt.plot(x, y)
plt.ylim(-1.0, 6.0)
```
