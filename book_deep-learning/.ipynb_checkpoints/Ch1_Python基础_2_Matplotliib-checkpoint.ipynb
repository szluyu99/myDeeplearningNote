### 绘制简单图形

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 生成数据
x = np.arange(0, 6, 0.1) # 以0，1为单位，生成0到6的数据，不包含0.6
y = np.sin(x)
```

这里使用NumPy的arange方法生成了[0, 0.1, 0.2, ..., 5.8, 5.9]的数据，将其设为x。对x的各个元素，应用NumPy的sin函数np.sin()，将x、y的数据传给plt.plot方法，然后绘制图形。

```python
# 绘制图形
plt.plot(x, y)
```

### pyplot的功能
在刚才的sin函数的图形中，我们尝试追加cos函数的图形，并尝试使用pyplot的添加标题和x轴标签名等其他功能。

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 生成数据
x = np.arange(0, 6, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
# 绘制图形
plt.plot(x, y1, label="sin")
plt.plot(x, y2, linestyle="--", label="cos") # 虚线绘制
plt.xlabel("x") # x轴标签
plt.ylabel("y") # y轴标签
plt.title('sin & cos') # 标题
plt.legend()
plt.show()
```
